# Multi-day backtest roll summary

Loads every `data/daily_news/*/processed/backtest_results.json` and prints aggregate stats.

**CLI equivalent:** run from repo root:

`python scripts/aggregate_backtest_results.py --output notebooks/_exported/backtest_roll_summary.json`


```python
from pathlib import Path
import json
import subprocess
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
script = ROOT / "scripts" / "aggregate_backtest_results.py"
out = ROOT / "notebooks" / "_exported" / "backtest_roll_summary.json"
out.parent.mkdir(parents=True, exist_ok=True)

subprocess.check_call([sys.executable, str(script), "--data-root", str(ROOT / "data" / "daily_news"), "--output", str(out)])
summary = json.loads(out.read_text(encoding="utf-8"))
summary
```
